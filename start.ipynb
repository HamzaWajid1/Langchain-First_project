Encoding the sentence to a vector having 768 parameters whose values will be dependent on the context of the text

```python
%pip install sentence_transformers
from sentence_transformers import SentenceTransformer

encoder=SentenceTransformer("all-mpnet-base-v2")
vectors=encoder.encode(df.text)
vectors.shape
```

```python
dim=vectors.shape[1]
dim
```

Using faceboom ai semantic search library to create index using flat L2.I should research about l2 later

```python
import faiss

index=faiss.IndexFlatL2(dim)
index
```

Adding the 8 vectors to the index and all vectors will have values that will values depending on their context

```python
index.add(vectors)
```

Writing a search query and then encoding it to a vector having 768 elements

```python
search_query="I want to invest in tesla stocks this month"

vec=encoder.encode(search_query)
vec.shape
```

Reshaping the vector into a desired format for search

```python
import numpy as np
svec=np.array(vec).reshape((1,vec.shape[0]))
svec.shape
```

Semantic Searching using search function.Search will be done by comparing the search query vector with the index vectors.The indexes of the 2 vectors have similar element values to the search query will be displayed

```python
distances,I=index.search(svec,k=2)
distances
```
